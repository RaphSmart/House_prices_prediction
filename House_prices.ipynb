```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the dataset
data_df = pd.read_csv("housing.csv")
data_df.head()
```

```python
# to check for null or missing values
data_df.info()
```

```python
# drop the null values
data_df.dropna(inplace = True)
```

```python
data_df.info()
```

```python
# visualize the dataset
data_df.hist(figsize = (15, 8))
```

```python
data_df.corr()
```

```python
# visualize the correlation with a heatmap
plt.figure(figsize = (15, 8))
sns.heatmap(data_df.corr(), annot = True, cmap = "YlGnBu")
```

```python
# view the values for categorical data (ocean_proximity)
data_df.ocean_proximity.value_counts()
```

```python
# one_hot_encoding with pandas
categorical_data = pd.get_dummies(data_df.ocean_proximity)
categorical_data.head()
```

```python
# drop 'ocean_proximity' that had been converted
data_df = data_df.drop(['ocean_proximity'], axis = 1)
data_df.head()
```

```python
# join one_hot coded column with data frame
data_final = data_df.join(categorical_data)
data_final.head()
```

```python
# visualize the house prices with a scatter plot
plt.figure(figsize = (15, 8))
sns.scatterplot(x = "latitude", y = "longitude", data = data_final, hue = "median_house_value", palette = "coolwarm")
```

```python
# drop median_house_value and assign to y for model evaluation
X = data_final.drop(["median_house_value"], axis = 1)
y = data_final["median_house_value"]
```

```python
X.head()
```

```python
y
```

```python
# split data into train and test and scale the data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

## Model

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

linReg = LinearRegression()
linReg.fit(X_train, y_train)
```

```python
y_pred = linReg.predict(X_test)
y_pred
```

```python
print("R2: ", r2_score(y_test, y_pred))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

RandForest = RandomForestRegressor()

RandForest.fit(X_train, y_train)
```

```python
RandForest.score(X_test, y_test)
```
